```python
import os
import pandas as pd
```

```python
for file in os.listdir():
    if file.split(".")[-1] not in ["xls", "xlsx"]:
        continue
    df = pd.read_excel(file)
    codes = dict()
    found = []
    for i in df.index:
        code = df["code"][i]
        if code != "unmatched entities":
            if codes.get(code, -1) != -1:
                if df["ps_id"][i] >= max(codes[code]) + 10:
                    found.append(code)
                codes[code].append(df["ps_id"][i])
            else:
                codes[code] = [df["ps_id"][i]]
    if len(found) > 0:
        print(file + ":")
        for e in found:
            print(str(e) + ": " + str(codes[e]))
        print("-")
        
    
```
