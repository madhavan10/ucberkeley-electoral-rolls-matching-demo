```python
import pandas as pd
import os
from collections import OrderedDict
```

```python
os.chdir("C:/users/madha/dropbox/electoral_rolls_matching/up_matchfiles/include_left/completed/codes")
os.listdir()
```

```python
df_stats = pd.DataFrame()
filenames = []
total_ps = []
unmatched_ps = []
some_unmatched_ps = []
unmatched_pct = []
some_unmatched_pct = []
for file in os.listdir():
    df_codes = pd.read_excel(file)
    Dict = OrderedDict()
    for i in df_codes.index:
        ps_id = df_codes["ps_id"][i]
        # ps_id not in dict
        if Dict.get(ps_id, -1) == -1:
            Dict[ps_id] = [df_codes["code"][i]]
        else:
            Dict[ps_id].append(df_codes["code"][i])
    rural_ps_total = len(Dict.keys())
    unmatched = 0
    some_unmatched = 0
    for k, v in Dict.items():
        if "unmatched entities" in v and len(v) == 1:
            unmatched += 1
        elif "unmatched entities" in v:
            some_unmatched += 1
    filenames.append(file)
    total_ps.append(rural_ps_total)
    unmatched_ps.append(unmatched)
    unmatched_pct.append(round(unmatched * 100.0 / rural_ps_total, 2))
    some_unmatched_ps.append(some_unmatched)
    some_unmatched_pct.append(round(some_unmatched * 100.0 / rural_ps_total, 2))
df_stats = pd.DataFrame({
    "File" : filenames, "Total # of polling stations(rural):": total_ps,\
    "# of unmatched polling stations": unmatched_ps, "% of unmatched polling stations": unmatched_pct,\
    "# of polling stations with some (but not all) names unmatched": some_unmatched_ps,
    "% of polling stations with some (but not all) names unmatched": some_unmatched_pct
})
df_stats.to_excel("C:/users/madha/dropbox/electoral_rolls_matching/up_stats.xlsx", index = False)
        
    
```
